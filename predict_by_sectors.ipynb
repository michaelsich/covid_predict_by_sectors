```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('geographic-sum-per-day-ver_00162.csv')
```

```python
jlm_df = df.loc[df['town_code'] == 3000]
jlm_df = jlm_df.drop(columns=['town_code', 'town'])
```

```python
sectors = pd.read_csv('stat_areas.csv')
sectors.columns = ['agas_code', 'heb_name', 'name']
sectors = sectors.drop(['heb_name'], axis=1)
```

```python
jlm_df = jlm_df.dropna(subset = ['agas_code'])
```

```python
jlm_df = jlm_df.replace(to_replace = "<15",value = 15)
jlm_df.dtypes
```

```python
jlm_df['accumulated_cases']= jlm_df['accumulated_cases'].astype('int')
jlm_df['accumulated_recoveries']= jlm_df['accumulated_recoveries'].astype('int')
jlm_df['accumulated_deaths']= jlm_df['accumulated_deaths'].astype('int')

jlm_df = jlm_df.assign(new_cases = jlm_df['accumulated_cases'] - jlm_df['accumulated_recoveries'] - jlm_df['accumulated_deaths'])
```

```python
jlm_df['accumulated_hospitalized']= jlm_df['accumulated_hospitalized'].astype('float')
jlm_df['accumulated_vaccination_first_dose']= jlm_df['accumulated_vaccination_first_dose'].astype('float')
jlm_df['accumulated_diagnostic_tests']= jlm_df['accumulated_diagnostic_tests'].astype('int')
jlm_df['accumulated_vaccination_second_dose']= jlm_df['accumulated_vaccination_second_dose'].astype('float')
jlm_df= jlm_df.sort_values(["agas_code","date"])
jlm_df.head()
```

```python
jlm_df["new_deaths"] = jlm_df.groupby("agas_code")["accumulated_deaths"].diff()
#jlm_df["new_cases_change"] = jlm_df.groupby("agas_code")["new_cases"].pct_change()
```

```python
jlm_df["new_recoveries"] = jlm_df.groupby("agas_code")["accumulated_deaths"].diff()
```

```python
jlm_df["new_deaths"] =jlm_df["new_deaths"].fillna(0)
jlm_df["new_recoveries"] =jlm_df["new_recoveries"].fillna(0)
jlm_df["new_cases_change"] =jlm_df["new_cases_change"].fillna(0)
```

```python
cases_date = jlm_df.pivot(index = 'agas_code',columns = 'date', values = 'new_cases')
cases_date
```

```python
deaths_date = jlm_df.pivot(index = 'agas_code',columns = 'date', values = 'new_deaths')
recoveries_date = jlm_df.pivot(index = 'agas_code',columns = 'date', values = 'new_recoveries')
```
